```python
## hlda
import tomotopy as tp
dataset = ['covid', '20', 'pubmed', 'yelp_korean', 'arxiv']
for model_path in dataset:
    model = tp.HLDAModel().load(f'../models/{model_path}/hlda.bin')
    num_leaf = 0
    for tn in range(model.k):
        if model.level(tn)==2: num_leaf+=1
    print(f'{model_path}: ',num_leaf)
```

```python
## cluhtm
def find_leaf_topics_cluhtm(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        line_count = 0
        lines_content = []
        for line in file:
            if line.count('\t') == 2:  # find the only leaf topic
                line_count += 1
                lines_content.append(line.strip())  
        return line_count, lines_content

file_path = '../models/20/hierarchical_structure_cluhtm.txt'

total_lines, lines_content = find_leaf_topics_cluhtm(file_path)

print(f"Num of leaf topics: {total_lines}")
print("Leaf topics:")
for line in lines_content:
    print(line)
```

```python
## hyhtm
def find_leaf_topics_hyhtm(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        line_count = 0
        lines_content = []
        for line in file:
            if line.startswith('\t\t'):  # find the leaf topics
                line_count += 1
                lines_content.append(line.strip())  
        return line_count, lines_content


file_path = '../models/yelp_korean/hierarchical_structure_hyhtm.txt'

total_lines, lines_content = find_leaf_topics_hyhtm(file_path)

print(f"Num of leaf topics: {total_lines}")
print("Leaf topics")
for line in lines_content:
    print(line)
```
