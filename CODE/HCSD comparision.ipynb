```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hlda = pd.read_feather('./result/hlda_result.ftr')
cluhtm=pd.read_feather('./result/cluhtm_result.ftr')
bertopic = pd.read_feather('./result/bertopic_result.ftr')
```

```python
phase = 1
hlda_result = hlda[hlda['phase']==phase]
cluhtm_result = cluhtm[cluhtm['phase']==phase]
bertopic_result = bertopic[bertopic['phase']==phase]

max_hcsd_hlda = hlda_result.loc[hlda_result['hcsd'].idxmax()]

max_hcsd_cluhtm = cluhtm_result.loc[cluhtm_result['hcsd'].idxmax()]

max_hcsd_bertopic = bertopic_result.loc[bertopic_result['hcsd'].idxmax()]

max_hcsd_result = max(max_hcsd_hlda, max_hcsd_cluhtm,max_hcsd_bertopic, key=lambda x: x['hcsd'])
```

```python
for i in [1,2,3,4]:
    phase = i
    hlda_result = hlda[hlda['phase']==phase]
    cluhtm_result = cluhtm[cluhtm['phase']==phase]
    bertopic_result = bertopic[bertopic['phase']==phase]

    max_hcsd_hlda = hlda_result.loc[hlda_result['hcsd'].idxmax()]

    max_hcsd_cluhtm = cluhtm_result.loc[cluhtm_result['hcsd'].idxmax()]

    max_hcsd_bertopic = bertopic_result.loc[bertopic_result['hcsd'].idxmax()]

    max_hcsd_result = max(max_hcsd_hlda, max_hcsd_cluhtm,max_hcsd_bertopic, key=lambda x: x['hcsd'])
    hlda_hcsd = hlda_result.hcsd
    cluhtm_hcsd = cluhtm_result.hcsd
    bertopic_hcsd = bertopic_result.hcsd

    max_hcsd_result=max_hcsd_result = max(max_hcsd_hlda, max_hcsd_cluhtm,max_hcsd_bertopic, key=lambda x: x['hcsd'])
    max_k = max_hcsd_result.k
    datafile = f'hcsd_data_{phase}.txt'
    with open(datafile, 'w') as f:
        for i in range(1, 10):
            f.write(f"{((i+1)*5)} {hlda_hcsd.values[i-1]} {cluhtm_hcsd.values[i-1]} {bertopic_hcsd.values[i-1]}\n")
            datafile = f'max_{phase}.txt'
    with open(datafile, 'w') as f:
        f.write(f"{max_hcsd_result.k} {max_hcsd_result.hcsd}\n")
```

```python

phase = 1
hlda_result = hlda[hlda['phase']==phase]
cluhtm_result = cluhtm[cluhtm['phase']==phase]
bertopic_result = bertopic[bertopic['phase']==phase]

max_hcsd_hlda = hlda_result.loc[hlda_result['hcsd'].idxmax()]

max_hcsd_cluhtm = cluhtm_result.loc[cluhtm_result['hcsd'].idxmax()]

max_hcsd_bertopic = bertopic_result.loc[bertopic_result['hcsd'].idxmax()]

max_hcsd_result = max(max_hcsd_hlda, max_hcsd_cluhtm,max_hcsd_bertopic, key=lambda x: x['hcsd'])




plt.figure(figsize = (20,10))
plt.title('Feb.2021-Jul.2021', fontsize =40)
plt.plot(hlda_result.k, hlda_result.hcsd, label = 'hLDA', linewidth=4)
plt.plot(cluhtm_result.k, cluhtm_result.hcsd, label = 'CluHTM', linewidth=4)
plt.plot(bertopic_result.k, bertopic_result.hcsd, label = 'BERTopic', linewidth=4)

plt.scatter(max_hcsd_result.k, max_hcsd_result.hcsd, marker='*', color='red', s=500, label='Max HCSD')
plt.axvline(x=max_hcsd_result.k, color='black', linestyle='--', linewidth=2)

#plt.ylim(0.45,0.55)
plt.xticks(fontsize =25)
plt.yticks(fontsize =25)
plt.xlabel('# Leaf Topics', fontsize =30)
plt.ylabel('HCSD score', fontsize =30)
plt.legend(fontsize=25,loc='upper right', bbox_to_anchor=(1.215, 1))
#plt.savefig(f'./result/hcsd_{phase}.png',bbox_inches='tight')
```
