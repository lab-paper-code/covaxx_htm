```python
## get best HARIN score in each dataset
news_max_harin_model = news_data.groupby('model')['HARIN'].idxmax()
news_max_harin_model= news_data.loc[news_max_harin_model]

covid_max_harin_model = covid_data.groupby('model')['HARIN'].idxmax()
covid_max_harin_model= covid_data.loc[covid_max_harin_model]

pubmed_max_harin_model = pubmed_data.groupby('model')['HARIN'].idxmax()
pubmed_max_harin_model= pubmed_data.loc[pubmed_max_harin_model]

arxiv_max_harin_model = arxiv_data.groupby('model')['HARIN'].idxmax()
arxiv_max_harin_model = arxiv_data.loc[arxiv_max_harin_model]

yelp_korean_max_harin_model = yelp_korean_data.groupby('model')['HARIN'].idxmax()
yelp_korean_max_harin_model = yelp_korean_data.loc[yelp_korean_max_harin_model]
```

```python
covid_max_harin_model['s+d'] = covid_max_harin_model['s+d']/2
covid_max_harin_model['s+c'] = covid_max_harin_model['s+c']/2
covid_max_harin_model['c+d'] = covid_max_harin_model['c+d']/2
covid_max_harin_model
```

```python
news_max_harin_model['s+d'] = news_max_harin_model['s+d']/2
news_max_harin_model['s+c'] = news_max_harin_model['s+c']/2
news_max_harin_model['c+d'] = news_max_harin_model['c+d']/2
news_max_harin_model
```

```python
pubmed_max_harin_model['s+d'] = pubmed_max_harin_model['s+d']/2
pubmed_max_harin_model['s+c'] = pubmed_max_harin_model['s+c']/2
pubmed_max_harin_model['c+d'] = pubmed_max_harin_model['c+d']/2
pubmed_max_harin_model
```

```python
yelp_korean_max_harin_model['s+d'] = yelp_korean_max_harin_model['s+d']/2
yelp_korean_max_harin_model['s+c'] = yelp_korean_max_harin_model['s+c']/2
yelp_korean_max_harin_model['d+c'] = yelp_korean_max_harin_model['d+c']/2
yelp_korean_max_harin_model
```

```python
arxiv_max_harin_model['s+d'] = arxiv_max_harin_model['s+d']/2
arxiv_max_harin_model['s+c'] = arxiv_max_harin_model['s+c']/2
arxiv_max_harin_model['c+d'] = arxiv_max_harin_model['c+d']/2
arxiv_max_harin_model
```

```python
# add Human Score from questionnaire
covid_max_harin_model['Human Score'] = [0.66,0.56,0.79]
news_max_harin_model['Human Score'] = [0.56,0.7,0.81,0.53]
pubmed_max_harin_model['Human Score'] = [0.69,0.59,0.88,0.45]
yelp_korean_max_harin_model['Human Score'] = [0.64, 0.48, 0.79, 0.36]
arxiv_max_harin_model['Human Score'] = [0.68, 0.6, 0.73, 0.43]
```

```python
print("covid\n")
print(covid_max_harin_model[['s', 'd', 'c','HARIN','Human Score']].corr())
print("\n\n")

print("20\n")
print(news_max_harin_model[['s', 'd', 'c','HARIN', 'Human Score']].corr())
print("\n\n")

print("pubmed\n")
print(pubmed_max_harin_model[['s', 'd', 'c','HARIN','Human Score']].corr())
print("\n\n")

print("Yelp\n")
print(yelp_korean_max_harin_model[['s', 'd', 'c','HARIN', 'Human Score']].corr())
print("\n\n")

print("arxiv\n")
print(arxiv_max_harin_model[['s', 'd', 'c','HARIN', 'Human Score']].corr())
```
