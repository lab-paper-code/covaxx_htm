```python
#  Calculate MRR based on mapping_dict and human_rating and output Reciprocal Rank for each dataset
def calculate_mrr(system_rankings, human_rating):
    reciprocal_ranks = {}
    
    for dataset, human_ranking in human_rating.items():
        system_ranking = system_rankings.get(dataset)
        
        if system_ranking is not None:
            for i, model_index in enumerate(system_ranking):
                if model_index in human_ranking:
                    rank = human_ranking.index(model_index) + 1  # Human ranking : (1-based)
                    reciprocal_ranks[dataset] = 1 / rank
                    break  
        else:
            reciprocal_ranks[dataset] = 0  
    
    # Reciprocal Rank 
    for dataset, rank in reciprocal_ranks.items():
        print(f"{dataset}: Reciprocal Rank = {rank}")
    
    # Mean Reciprocal Rank : (MRR)
    if reciprocal_ranks:
        return sum(reciprocal_ranks.values()) / len(reciprocal_ranks)
    else:
        return 0  



# MRR Calculation
print('sim')
sim_mrr = calculate_mrr(similarity_rating, human_rating)
print(f"\nMRR (Similarity): {sim_mrr}")
print('\n')

print('div')
div_mrr = calculate_mrr(diversity_rating, human_rating)
print(f"\nMRR (Diversity): {div_mrr}")
print('\n')

print('coh')
coh_mrr = calculate_mrr(coherence_rating, human_rating)
print(f"\nMRR (Coherence): {coh_mrr}")
print('\n')

print('HARIN')
HARIN_mrr = calculate_mrr(HARIN_rating, human_rating)
print(f"\nMRR (HARIN): {HARIN_mrr}")
print('\n')

print('S+D')
sim_div_mrr = calculate_mrr(s_d_rating, human_rating)
print(f"\nMRR (S+D): {sim_div_mrr}")
print('\n')

print('D+C')
div_coh_mrr = calculate_mrr(d_c_rating, human_rating)
print(f"\nMRR (D+C): {div_coh_mrr}")
print('\n')

print('S+C')
sim_coh_mrr = calculate_mrr(s_c_rating, human_rating)
print(f"\nMRR (S+C): {sim_coh_mrr}")
print('\n')



```
