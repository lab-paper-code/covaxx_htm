```python
import pandas as pd
import random
```

```python
data = pd.read_feather('../Datasets/dataset.ftr')
```

```python
for i in range(5):
    sample = data.sample(frac=0.1)
    sample.reset_index(drop=True).to_feather(f'../Datasets/sampledf_{i}.ftr')
    file = open(f'../Datasets/sampletxt_{i}.txt','w')
    for row in sample['okt']:
        file.write(' '.join(row)+'\n')
```
