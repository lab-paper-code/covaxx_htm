```python
import openai
from googletrans import Translator
```

```python
openai.api_key = "api_key"
```

```python
words = ['menstruation',' side effects', 'bleeding', 'menstrual cycle', 'one week'
]
```

```python
def generate_title(words):
    prompt = f"associated with COVID-19 Vaccine, select the most representative one from {words}"
    print(prompt)
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=prompt,
        max_tokens=20
    )
    title = response.choices[0].text.strip()
    return title
```

```python
generated_title = generate_title(words)
print(f"Generated Title: {generated_title}")
```
