```python
import pandas as pd
from wordcloud import WordCloud
from collections import Counter
from googletrans import Translator
```

```python
def translate_to_english(korean_word):
    translator = Translator()

    return translator.translate(korean_word, src="ko", dest="en").text
    
```

```python
data=pd.read_feather('../Datasets/dataset.ftr')
vocab = []
for line in data.okt:
    for word in line:
        vocab.append(word)

        
```

```python
counts = Counter(vocab)
tags = counts.most_common(50) 
```

```python
new_tags = {}
for t in tags:
    new_tags[translate_to_english(t[0])] = t[1]
```

```python
wc = WordCloud(background_color="white", max_font_size=60)
cloud = wc.generate_from_frequencies(new_tags)


# 생성된 WordCloud를 test.jpg로 보낸다.
cloud.to_file('../figures/wordcloud.pdf')
```
