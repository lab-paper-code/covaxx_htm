```python
def get_standard_dictionary(word):
    url = "https://stdict.korean.go.kr/api/search.do"
    params = {
        "key": "api_key", # 국립국어원에서 API키를 발급받아서 입력하세요
        "q": word,
        "req_type": "json",
    }
    response = requests.get(url, params=params)
    response.json(encoding = 'UTF-8')
    return response
```

```python
data = pd.read_feather('../Datasets/all.ftr')
```

```python
data
```

```python
vocab = set([])
for i in range(len(data)):
    vocab = vocab.union(set(data.okt[i]))
```

```python
vocab = list(vocab)
```

```python
stop_words = []
for i in tqdm.tqdm(range(len(vocab))):
    try: 
        t = get_standard_dictionary(vocab[i])
        key = 0
        for temp in t.json().get('channel').get('item'):
            if temp.get('pos') == '명사' or temp.get('pos') == '형용사':
                key =1
        if key==0 : stop_words.append(vocab[i])        
    
    except: 
        stop_words.append(vocab[i])
```

```python
user_dict_file = "../Datasets/RawDatasets/txt/user_dic.txt"
user_dict =  [line.strip() for line in open(user_dict_file, encoding="utf-8").readlines()]
```

```python
f = open("../Datasets/RawDatasets/txt/new_stopwords.txt",'w')
```

```python
for word in stop_words:
    if word not in user_dict:
        f.write(word+"\n")
    
```
